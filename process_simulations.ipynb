```python
# -- FUNCTION TO GET THE MODIFIED PARAMETER VALUE --
#
def get_modified_value(df_base_param, df_param):

    df_base_param_keys   = df_base_param.columns
    df_base_param_values = df_base_param.to_numpy()[0]
    df_param_keys        = df_param.columns
    df_param_values      = df_param.to_numpy()[0]

    key_params           = ["radiality", "EV", "EHP", "storage", "Storage_cost", "network_reconfig", "bilevel", "PV_CAPA", "PVC","DSOEC","EIC","EEC","GCC", "weight_I"]
  
    for (i, (v1, v2)) in enumerate(zip(df_param_values, df_base_param_values)):
        key = df_param_keys[i]
        if  key in key_params and v1 != v2:
            if key == "storage":
                value_changed_param = df_param_values[df_param_keys == "Storage_cost"][0]
                key_changed_param   = "Storage_cost"
            else:
                value_changed_param = v1
                key_changed_param   = key

            return key_changed_param, value_changed_param

    return None 
```

```python
# -- FETCHING THE DATA FROM THE EXCEL FILES --
# Path
cwd       = Path.cwd()
simu_path = Path(cwd, 'simulations')
# Constant
simu_replicates = [2, 7, 9, 11, 15, 18, 21, 24, 27, 30]
simu_indices    = [i for i in range(0,33) if i not in simu_replicates]
#print(simu_indices)


# Constant
directories = list()
for name in ["single_commodity", "multi_commodity"]:
    for i in simu_indices:
        directory = "simu_"+name+"_"+str(i)
        temp_dir = Path(simu_path, directory)
        if temp_dir.is_dir():
            directories.append(temp_dir)

excel_files = [Path(d, 'results.xlsx') for d in directories]

df_param, df_results = list(), list()
for i, e in enumerate(excel_files):
    df_param.append(pd.read_excel(e, index_col=0, header=None, skiprows=[0]+[i for i in range(34,112)], usecols=[0,1]).transpose())
    temp_xlsx = (pd.read_excel(e, index_col=0, header=None, usecols=[3,4,5]).transpose())
    if temp_xlsx.columns[0] == "failure":
        temp_xlsx = pd.DataFrame(columns = [all_KPIs, all_units], data = np.reshape(["-" for i in range(0, len(all_KPIs))], (1,111)))
        df_results.append(temp_xlsx)
    else:
        temp_xlsx.fillna('-', inplace=True)
        temp_xlsx.columns = [all_KPIs, all_units]
        temp_xlsx.keys    = [all_KPIs, all_units]
        temp_xlsx.data    = temp_xlsx.values[1] 

        # Retreives the index of the first line with the bullshit
        index_bullshit = temp_xlsx.index[0]

        # Send to hell the bullshit
        temp_xlsx = temp_xlsx.drop([index_bullshit])

        # Append the clean data :-)
        df_results.append(temp_xlsx)

# Concatenation of all the tables
key_params_tab = pd.concat(df_param, axis=0)[key_params]
results_tab    = pd.concat(df_results, axis=0, ignore_index=False)

# Displaying the good ass looking table
display(results_tab)

# Definition of the base cases
df_param_basecase_single   = df_param[0].copy(deep=True)
# df_results_basecase_single = df_results[0].copy(deep=True)
df_param_basecase_multi    = df_param[len(simu_indices)].copy(deep=True)
# df_results_basecase_multi  = df_results[len(simu_indices)].copy(deep=True)
```

```python
def stackBarPlot(df_violation: pd.DataFrame,
                 BGV: list=['Study_Case', 'Violation_Class','mFRR_max'],
                 subplot_var: str='Scenario',
                 fig_size: tuple=(1800,800),
                 fontsize: float=22
                ):
    """Plots a grouped stacked bar plot with at most 3 bars per group.

        Parameters
        ----------
        df_violation: pd_DataFrame
            Dataframe containing the violation durations to plot.
        BGV: list
            List containing three strings:
                - String 0: Bar variable
                - String 1: the group variable
                - String 2: the stack variable
        subplot_var: string
            String indicating on which datframe column we want to separate
            the plots
        fig_size: tuple
            Tuple=(width,height)
        fontsize: float
            Fontsize for the texts written in the figure
        Return
        ----------
        fig: a plotly figure object
            The grouped stack bar plot figure
        """
    ###########################################################################
    # Check inputs
    ###########################################################################
    if len(BGV) != 3:
        raise Exception("BGV must contain 3 elements")
    if len(df_violation[BGV[0]].unique()) > 3:
        raise Exception("Max 3 different values for the bar variable")

    ###########################################################################
    # Inputs
    ###########################################################################
    unity = {
            'mFRR_max': 'MW', 
            'Violation_Class': '-',
            'Violation_duration': 'min',
            'Study_Case': '-'}
    # BGV input = B: bar, G: group, V: vertical
    bar_variable = BGV[0]
    group_variable = BGV[2]
    stack_variable = BGV[1]
   
    # Change type of mFRR reserve to have it without decimals
    df_violation['mFRR_max'] = df_violation['mFRR_max'].astype('int')

    # Unique values of the input variables in the dataframe
    bar_values = df_violation[bar_variable].unique()
    group_values = df_violation[group_variable].unique()
    stack_values = df_violation[stack_variable].unique()
    subplot_var_values = df_violation[subplot_var].unique()

    # Max 3 different bars to compare so 3 different colors max
    colors = []

    colors.append(col.n_colors('rgb(255, 229, 191)','rgb(175, 79, 35)', #orange
                  n_colors=len(stack_values), colortype='rgb')) 
    colors.append(col.n_colors('rgb(199, 232, 252)','rgb(2, 73, 140)',  #blue
                  n_colors=len(stack_values), colortype='rgb')) 
    colors.append(col.n_colors('rgb(216,243,220)','rgb(27,67,50)', #green
                  n_colors=len(stack_values), colortype='rgb')) 

    """
    colors.append(['rgb(255, 229, 191)','rgb(255, 229, 191)']) #orange
              
    colors.append(['rgb(199, 232, 252)','rgb(199, 232, 252)'])  #blue
    """
    ###########################################################################
    # Creation of the grouped stack bar plot figure
    ###########################################################################

    subplot_titles = [x.replace('_', ' ') for x in subplot_var_values]
    fig = make_subplots( 
                        rows=1, 
                        cols=len(subplot_var_values), 
                        subplot_titles=subplot_titles,
                        column_widths=[0.5, 0.5],
                        horizontal_spacing=0.02
                        )

    for (i, sub) in enumerate(subplot_var_values): # iterate over the subplots
        temp = df_violation[df_violation[subplot_var]==sub] 
        for (j, bar) in enumerate(bar_values): # iterate over the bars
            temp1 = temp[temp[bar_variable]==bar]
            v = -1
            previous_x = 0
            for (k, stack) in enumerate(stack_values): # iterate over the stacks
                temp2 = temp1[temp1[stack_variable]==stack]
            
                if v != -1:
                    previous_x += current_x
                current_x = temp2['Violation_Duration'].values
                
                if i == len(subplot_var_values)-1:
                    showlegend = True
                else: 
                    showlegend = False

                current_y = temp2[group_variable].astype('str')
                
                fig.add_trace(go.Bar(
                                name="%s: %s %s" %(stack_variable.replace('_',' '), stack,  unity[stack_variable].replace('-','')),
                                x=current_x,
                                y=current_y,
                                legendgroup="group%s"%j,
                                legendgrouptitle_text="%s: %s" %(bar_variable.replace('_',' '), bar),
                                offsetgroup=j,
                                orientation='h',
                                base=previous_x,
                                marker_color=colors[j][k],
                                showlegend=showlegend,
                                #text=current_x.round(decimals=0),
                                legendrank = len(bar_values)-j
                                ), row = 1, col=i+1
                            )
                
                v += 1
```

```python
import math
import seaborn as sns

# Color palette
colors_smart = sns.color_palette("Paired", 23)

# -- Create plots --
grid_h = len(df_key_KPIs_single.columns)
grid_w = 1
fig = plt.figure(figsize = (13, 20))
gs  = gridspec.GridSpec(grid_h, grid_w, wspace = 0.05, hspace = 0.3, figure = fig)

factor = 5

# All the simulation indexes
index_simu = list(df_key_KPIs_single.index)

# Looping over each plots
for line, kpi in enumerate(df_key_KPIs_single.columns):
        
    # Creation of subplot
    ax = plt.subplot(gs[line, 0])
    spacing = 0.5
    # plt.subplots_adjust(bottom=spacing)
    plt.xlabel("Simulation id", fontsize = 12)
 
    # Fixing y scale
    # ax.set_ylim(bottom = -0.6, top = 1.1)            
    # ax.tick_params(axis = 'y', labelsize = 10)

    # Removing x-labels
    ax.set_xticks(df_key_KPIs_single.index)
    ax.label_outer()
    
    # Values
    values = df_key_KPIs_single[kpi]

    vmax = 0
    vmin = 0

    # Looping over all the simulations
    for index_sim, val in enumerate(values):

        # FAILURE
        if val == "-" or val == -1000:
            val = 0
        
        # Making sure its a float
        val = float(val)
            
        # Plotting
        bars = ax.bar(index_sim, val, color = colors_smart[index_sim])

        # Updating y limits
        vmax = max(vmax, val)
        vmin = min(vmin, val)

        # Beautifier
        beau_name = "X" if val == 0.0 else round(val, 3)

        # Adding a nice ass looking text on top of the bar
        for bar in bars:
            height = bar.get_height()
            ax.annotate(beau_name, xy=(bar.get_x() + bar.get_width() / 2, height), xytext=(0, 1),
            textcoords="offset points", ha='center', va='bottom', fontsize = 7)  

        ax.set_ylim(bottom = vmin, top = vmax * (1 + 1/factor))
        ax.grid(visible=False, axis='x')
        ax.grid(visible=True, axis='y',alpha=0.5)

    # Adding title with term
    plt.ylabel(kpi[0]+ f" [{kpi[1]}]", fontsize = 10)
    #spacing = 0.100
    #.subplots_adjust(bottom=spacing)

plt.savefig(f"kpi.pdf", bbox_inches='tight')
```
